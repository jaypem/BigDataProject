```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime as dt
```

```python
data = pd.read_csv('output.txt', header=None, names=['Date', 'Start', 'End', 'Negative', 'Neutral', 'Positive'])
```

```python
data = data.sort_values(['Date', 'Start'])
data.head()
```

```python
data['Sum'] = data[['Negative', 'Neutral', 'Positive']].sum(axis=1)
data.head()
```

```python
bitcoin = pd.read_csv('Bitcoin.csv', sep=';', decimal=',')
bitcoin
```

```python
bc = np.zeros(len(data))

for i in range(len(data)): 
    start_d = dt.datetime.strptime(data.iloc[i]['Date'] + ' ' + str(data.iloc[i]['Start']) + ':00', '%Y-%m-%d %H:%M')
    
    if data.iloc[i]['End'] == 24:
        end_d = dt.datetime.strptime(data.iloc[i]['Date'] + ' 00:00', '%Y-%m-%d %H:%M')
        end_d += dt.timedelta(days=1)
    else:
        end_d = dt.datetime.strptime(data.iloc[i]['Date'] + ' ' + str(data.iloc[i]['End']) + ':00', '%Y-%m-%d %H:%M')
    
    start_key = dt.datetime.strftime(start_d, '%Y-%m-%d %H:%M')
    end_key = dt.datetime.strftime(end_d, '%Y-%m-%d %H:%M')
    
    o = bitcoin[bitcoin['Date'] == start_key]['Open Price'].values[0]
    c = bitcoin[bitcoin['Date'] == end_key]['Close Price'].values[0]
    
    bc[i] = c - o
```

```python
neg = (data['Negative'] / data['Sum']).values
neu = (data['Neutral'] / data['Sum']).values
pos = (data['Positive'] / data['Sum']).values

x_ticks = (data['Date'] + "\n" + data['Start'].astype(str) + "-" + data['End'].astype(str) + 'h').values

width = 1
ind = np.arange(len(data))

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()

ax1.set_xlabel('time')

ax1.bar(ind, neg, width, color='r')
ax1.bar(ind, neu, width, bottom=neg, color='gold')
ax1.bar(ind, pos, width, bottom=neg+neu, color='g')

ax1.set_ylabel('sentiments')

ax2.plot(ind, bc, 'b-', zorder=10)
ax2.plot(ax1.get_xlim(), [0, 0], 'k--')

m = max(abs(bc)) + 10
ax2.set_ylim([-m, m])

ax2.set_ylabel('bitcoin trend')
ax2.tick_params('y', colors='b')

plt.xticks(ind, x_ticks)

plt.show()
```

```python
neg = data['Negative'].values
neu = data['Neutral'].values
pos = data['Positive'].values

x_ticks = (data['Date'] + "\n" + data['Start'].astype(str) + "-" + data['End'].astype(str) + 'h').values

width = 1
ind = np.arange(len(data))

fig, ax1 = plt.subplots(figsize=(20, 10))

ax2 = ax1.twinx()

c = 16000
ax1.set_ylim([0, c])
centers = c / 2 - neg - neu / 2

ax1.set_xlabel('time')

ax1.bar(ind, centers, width, color='w')
ax1.bar(ind, neg, width, bottom=centers, color='r')
ax1.bar(ind, neu, width, bottom=centers + neg, color='gold')
ax1.bar(ind, pos, width, bottom=centers + neg + neu, color='g')

ax1.set_ylabel('sentiments', fontsize=40)
ax1.get_yaxis().set_ticks([])
ax1.legend(['Sentiments','negative', 'neutral', 'positive'], fontsize=18)

ax2.plot(ind, bc, 'b-')
ax2.plot(ax1.get_xlim(), [0, 0], 'k--')

m = max(abs(bc)) + 10
ax2.set_ylim([-m, m])

ax2.set_ylabel('bitcoin trend', fontsize=40)
ax2.tick_params('y', colors='b', labelsize=20)

plt.xticks(ind, x_ticks, fontsize=40)

#plt.show()
plt.savefig('result.png')
```
